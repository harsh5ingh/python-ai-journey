```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data Structures: [resturant_id, 2021, 2022, 2023, 2024]

```python
sales_data = np.array([
  [1, 150000, 180000, 220000, 250000], # Paradise Briyani
  [2, 120000, 140000, 160000, 190000], # Beijing Bites
  [3, 200000, 230000, 260000, 300000], # Pizza Hub
  [4, 180000, 210000, 240000, 270000], # Burger Point
  [5, 160000, 185000, 205000, 230000], # Chai Point
])

print("=== Zomato sales analysis === ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restaurants: ", sales_data[:3]) # [0:3] or [:3] same h.
```

### Total sales per year

```python
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

# Minimum sales per restaurants
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# Maximum sales per year

max_sales = np.max(sales_data[:, 1:], axis= 0)
print(max_sales)

# Average sales per restaurant

avg_Sales = np.average(sales_data[:, 1:], axis=1)
print(avg_Sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize= (8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restaurant ")
plt.xlabel("Years ")
plt.ylabel("Sales ")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10 ])

print("Vector addition: ", vector1 + vector2)
print("\n Multiplication: ", vector1 * vector2)
print("\n Dot product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

### Vectorized operations

```python
restuarants_types = np.array(['briyani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper", vectorized_upper(restuarants_types))
```
