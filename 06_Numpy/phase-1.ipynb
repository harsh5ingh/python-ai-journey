## Numpy array and basics
learn about numpy

```python
import numpy as np
```

### creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print("1D array", arr_1d)

arr_2d = np.array([[1,2,3], [4,5,6]])
print("2D array", arr_2d)
```

### List vs numpy array

```python
py_list = [1,2,3]
print("Python list multiplication ", py_list*2)


np_array = np.array([1,2,3]) #element wise multiplication
print("Python array multiplication ", np_array*2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time: ", time.time() - start)

start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy operation time: ", time.time() - start)
```

### creating array from scratch

```python
zeros = np.zeros((3, 4))
print("zeros array:  \n", zeros)

ones = np.ones((2, 3))
print("ones array:  \n", ones)

full = np.full((2, 2), 7)
print("full array:  \n", full)

random = np.random.random((2,3))
print("random array:  \n", random)

sequence = np.arange(0, 10, 2) # 0 to 10 ke bich 2-2 ke gap pe jayega
print("sequence array:  \n", sequence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1,2,3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3],
                  [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1,2], [3, 4]],
                   [[5, 6], [7, 8]]]) # tensor m dimension jyada hota h
print("Tensor: ", tensor)
```

### Array properties

```python
arr = np.array([[1, 2 , 3],
                [4, 5, 6]])
print("Shape ", arr.shape) # shape find
print("Dimension ", arr.ndim) #dimension find
print("Size ", arr.size) # size means how much elements are there
print("DType ", arr.dtype) # to find data type
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original array ", arr)

reshaped = arr.reshape((3, 4))
print("\n Reshape array ", reshaped)

flattened = reshaped.flatten()
print("\n Flatten array ", flattened)

# ravel( returns view, instead of copy )
raveled = reshaped.ravel()
print("\n Raveled array ", raveled)

# Transpose

transpose = reshaped.T
print("\n Transposed array ", transpose)
```
